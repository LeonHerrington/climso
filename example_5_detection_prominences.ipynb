# Example : Prominences

### Imports

```python
import os, re

from climso.detection import Prominences, ProminenceSpeed
```

### Files

```python
base_dir='example_data/CLIMSO/2023/data_calibrated/2023-07-28/' #2024-08-09
files = [base_dir+file for file in os.listdir(base_dir) if file.startswith('imoa_06563_c1_')]

def extract_timestamp(file_path):
    match = re.search(r'\d{8}_\d{8}', file_path)
    if match:
        return match.group(0)
    return ''

files = sorted(files, key=extract_timestamp)
```

```python
images = []
for file in files:
    images.append(Prominences(file))
```

Sort by date.

```python
images.sort(key=lambda x: x.date_obs)
```

### Single image prominences positions heights

```python
images[-5].plot()
```

```python
images[-5].prominences
```

### Prominence height and speed over time

```python
prom_speed = ProminenceSpeed(images, prominence_index=3)
```

```python
prom_speed.plotDistances()
```
