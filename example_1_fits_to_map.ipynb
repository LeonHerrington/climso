### SunPy GenericMap

**Sunpy GenericMap :** https://docs.sunpy.org/en/stable/generated/api/sunpy.map.GenericMap.html#sunpy.map.GenericMap

This function converts a fits, fts or fts.bz2 file into a SunPy GenericMap. SunPy provides many useful tools, so converting to a sunpy map is very beneficial. 

*The disk is supposed to be already centered during calibration however that is not always the case. It is recommended to apply the disk centering in this method, this also recalculates the solar radius that will be in the header (meta) of the GenericMap.*

```python
file = 'example_data/CLIMSO/2024/data_calibrated/2024-07-19/imoa_03933_l2_20240719_15340779_emi1.fts.bz2'
```

```python
map = toSunpyMap(file, center_disk=True)
```

```python
map.plot()
```
