# Example : filament detection

### Imports

```python
from climso.detection import Filaments
```

### File

```python
file = 'example_data/CLIMSO/2024/data_calibrated/2024-06-21/imoa_06563_l1_20240621_05522700_emi1.fts.bz2'
```

### Get filaments from file

```python
filaments = Filaments(file)
```

### Plot filaments

```python
filaments.plot(figsize=(6,6))
```
