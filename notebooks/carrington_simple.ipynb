# Carte Carrington simple

Exemple simple de "carte de carrington" sans projection

```python
import matplotlib.pyplot as plt
import numpy as np
import glob

from IPython.display import clear_output

import sys
sys.path.insert(0, '..')
from module import readFitsBz2
```

```python
path = 'Z:/data/CLIMSO/2024/data_calibrated/2024-03-19/'

data = []
counter = 0

files = glob.glob(path + '*_l2_*.fts.bz2')
total = len(files)
for file in files:
    data.append(readFitsBz2(file))
    counter+=1
    clear_output(wait=False)
    print(f'{counter}/{total}')
```

```python
carrington = data[0][:, 1024:1025]
for image in data[1:]:
    carrington = np.hstack((carrington, image[:,1000:1001]))
```

```python
plt.figure()
plt.imshow(carrington, cmap='grey');
plt.title('Carrington');
```
