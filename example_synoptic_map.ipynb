# Example : Synoptic Map

### Imports

```python
from climso.maps import SynopticMap
```

### Generate map

**climso_dir :** the directory containing the CLIMSO data (ex:'/data/CLIMSO/'). Assuming that the data is in the same structure as in climso-db-calc1.irap.omp.eu.

**rotation :** carrington rotation number of the synoptic map. SunPy provides a way of getting the carrington rotation number given a datetime or similar type object using the following function *sunpy.coordinates.sun.carrington_rotation_number(t='now')*. https://docs.sunpy.org/en/stable/generated/api/sunpy.coordinates.sun.carrington_rotation_number.html

**disk :** type of disk image to use to produce the maps. Either 'l1' (H alpha 656.3 nm) or 'l2' (Ca II 393.4nm).

```python
climso_dir = 'example_data/CLIMSO'
rotation = 2286
disk = 'l2'

synoptic_map = SynopticMap(climso_dir=climso_dir, rotation=rotation, disk=disk)
```

### Plot

Mostly just a matplotlib.pyplot.imshow(), but simplifies the task of setting ticks and labels. figsize, cmap, vmin and vmax are parameters of this method.

```python
synoptic_map.plot(cmap='grey')
```

```python
synoptic_map.plot(cmap='hinodesotintensity')
```
