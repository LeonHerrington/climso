# Example : Active Region Detection

### Imports

```python
from climso.detection import ActiveRegions
from climso.utils import toSunpyMap
```

### Convert to SunPy map

```python
file = 'example_data/CLIMSO/2024/data_calibrated/2024-07-19/imoa_03933_l2_20240719_15340779_emi1.fts.bz2'
```

```python
map = toSunpyMap(file, center_disk=True)
```

### Get active regions

```python
active_regions = ActiveRegions(map)
```

### Plot

```python
active_regions.plot(figsize=(6,6))
```

### Table

Table conataining active regions positions (longitude, latitude and pixel position), and filament surface area in uSH.

```python
active_regions.table
```
